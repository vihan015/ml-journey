```python
np.where(mask,True,False)
# np.where(condition) : returns the indices where the condition is true
# np.where(condition, x, y) : returns x if condition true and y if false
"""
if(condition) {
    x
} else {
    y
} 
"""
```

```python
# adding row and coloum in matrix
a = np.array([[1,2], [3,4]])
# vstack : vertical stack ( to add a new row)
# hstack : horizontal stack ( to add a new coloum)
new_row = np.array([[5,6]])
a = np.vstack((a,new_row))
print(a)
new_col = np.array([[7],[8],[9]])
a = np.hstack((a,new_col))
print("\n", a)
```
